```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
gamma_1 = 0
gamma_2 = 0
rho = 2
lamb = 1
def p(t):
    return [1, -(gamma_1 - 1j*rho + gamma_2 + 2*lamb*t), (gamma_1 - 1j*rho) * (gamma_2 + 2 * lamb * t) - lamb**(-2)]
```

```python
T_0 = -3
T_1 = 3
dt = 0.001
t = np.arange(T_0, T_1, dt)
```

```python
z = np.zeros(shape=(t.shape[0], 2), dtype=complex)
for i in range(len(t)):
    z[i] = np.roots(p(t[i]))
```

```python
plt.scatter(z[:,0].real, z[:,0].imag, marker='.', c=t, cmap='jet')
plt.scatter(z[:,1].real, z[:,1].imag, marker='.', c=t, cmap='jet')
plt.show()
```
